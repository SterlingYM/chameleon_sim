Difference from ```Chameleon 3D Simulation (I)```:

- No boundary condition for the inner sphere. This enables the simulation for weaker coupling where the value at the boundary is not at the equilibrium value. This allows us to simulate more complicated case (e.g. gaussian density distribution for the Cs source) as well.
- Slower simulation. The solution tends to blow up when random value is given as an initial guess. To prevent this, we give the inner equilibrium value as an initial guess, which allows us to get the proper solution with a cost of simulation running time.

````python
import fenics as fn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

## Parameters

````python
g_per_cm3_to_ev4 = 4.3 * 10**18
scaling = 1.97 * 10**-4
````

````python
Lambda = 0.1 * 10**-3 # [eV]
Mpl    = 2.4 * 10**27 # [eV]
M      = 10**-3 * Mpl
````

````python
R_sph = 10 # radius of internal sphere
````

````python
rho_inner = 2.7 * g_per_cm3_to_ev4 
rho_outer = 7   * g_per_cm3_to_ev4 
rho_vacuum = 6.6 * 10**-17  * g_per_cm3_to_ev4

inner_phi_eq = np.sqrt(Lambda**5 * M / rho_inner)
outer_phi_eq = np.sqrt(Lambda**5 * M / rho_outer)
````

## Functions & Classes

In this example, we define a class ```Rho``` which defines the density at each point in mesh. This method can be implemented either by defining two domains in mesh physical region file, or by manually setting the coordinate to re-map the domain in mesh (which is done by ```Inner()``` function in this case).

````python
def solve(V,rho_solve,phi):
    v = fn.TestFunction(V)
    f = Lambda**5/phi**2
    g = fn.Constant(0)
    a = (fn.inner(fn.grad(phi)*scaling**2,fn.grad(v))+ rho_solve*v/M)/scaling*fn.dx
    L = f*v/scaling*fn.dx + g*v*fn.ds
    F = a-L
    fn.solve(F==0, phi, bcs)#, solver_parameters={'newton_solver' : {'linear_solver' : 'mumps'}})
    return phi
````

````python
def importMesh(xml_file):
    physical_file = xml_file[:-4]+'_physical_region.xml'
    facet_file    = xml_file[:-4]+'_facet_region.xml'
    mesh       = fn.Mesh(xml_file)
    markers    = fn.MeshFunction("size_t", mesh, physical_file)
    boundaries = fn.MeshFunction('size_t', mesh, facet_file)
    dx = fn.Measure('dx', domain=mesh, subdomain_data=markers)
    V = fn.FunctionSpace(mesh, 'CG', 1)
    return mesh,markers,boundaries,dx,V
````

````python
class Inner(fn.SubDomain):
    tol = 1e-14
    def inside(self,x,on_boundary):
        return x[0]**2 + x[1]**2 + x[2]**2 <= (R_sph + self.tol)**2

class Rho(fn.UserExpression):
    def __init__(self,materials,rho_0,rho_1,**kwargs):
        super().__init__(**kwargs)
        self.materials = materials
        self.rho_0 = rho_0
        self.rho_1 = rho_1
    def value_shape(self):
        return ()
    def eval_cell(self,values,x,cell):
        if self.materials[cell.index] == 0:
            values[0] = self.rho_0
        else:
            values[0] = self.rho_1
````

````python
# prepare for plotting
x = np.linspace(-49.5,49.5, 1000)
y = np.zeros(len(x))
z = np.zeros(len(x))
coords = np.dstack((x, y, z))[0]
cm = plt.get_cmap('winter_r')
````

# main

````python
####  (1) Import mesh ####
mesh,markers,boundaries,dx,V = importMesh('models/sphere4.xml')
````

````python
#### (2) Boundary condition ####
outer_phi_eq = np.sqrt(Lambda**5 * M / rho_outer)
outer_boundary = fn.DirichletBC(V, fn.Constant(outer_phi_eq), boundaries, 1)
bcs = outer_boundary
````

````python
#### (3) set domains (for different densities) ####
domains = fn.MeshFunction('size_t',mesh,mesh.topology().dim())
domains.set_all(1)
inner = Inner()
inner.mark(domains,0)
````

````python
#### (4) solve pde (single run) ####
phi = fn.project(fn.Constant(inner_phi_eq),V)
rho = Rho(domains,rho_inner,rho_vacuum,degree=0)
phi = solve(V,rho,phi)
````

````python
data1D = np.array(list(map(phi,coords)))
plt.figure(figsize=(8,5))
plt.title('Chameleon Field')
plt.xlabel('r [mm]')
plt.ylabel(r'$\phi$ [meV]')
plt.plot(x,data1D*1000)
plt.axvspan(-10,10,label='Location of sphere',color='orange',alpha=0.2)
plt.legend()
plt.show()
````

**This result is consistent with plot (IV - 2) in ```Chameleon 3D Simulation I```.**

A few notes:

- It is obvious that the shape near the source surface depends on the mesh size. <br>
**TO ANALYZE THE FIELD NEAR THE SURFACE ACCURATELY, SIZE OF MESH MUST BE SMALLER THAN EXPECTED COMPTON LENGTH!!**<br>
This of course comes with the cost of large computing time.<br>

- As mentioned above, the solution does not converge properly if the initial guess is NOT a uniformly flat field at the expected equilibrium value. This is due to the property of the solver which is good at finding the nonzero curvature and is not good at choosing the uniform field as the $\nabla^2=0$ solution.

### Varying the density

Just for fun, we can vary the density to see the effect of coupling.

````python
#### (4) solve pde (test to vary rho value) ####
rho_list = rho_inner*np.logspace(-12,0,12)
plot_list = []
for i in range(len(rho_list)):
    print("\rSimulating: ({}/{})".format(i+1,len(rho_list)),end="")
    
    # start with expected equilibrium value inside the inner sphere
    # (otherwise the solution does not converge)
    inner_phi_eq = np.sqrt(Lambda**5 * M / rho_list[i])
    phi = fn.project(fn.Constant(inner_phi_eq),V)
    rho = Rho(domains,rho_list[i],rho_vacuum,degree=0)
    
    # solve
    phi = solve(V,rho,phi)
    data1d = np.array(list(map(phi,coords)))
    plot_list.append(data1d)
````

````python
# finalize the plot
plt.figure(figsize=(10,6))
for i in range(len(plot_list)):
    plt.plot(x,plot_list[i]*1000,label=r'$\rho$ = {:.2e} $eV^4$'.format(rho_list[i]),color = cm(1.*i/len(rho_list)))    
plt.xlabel('r [mm]')
plt.ylabel(r'$\phi$ [meV]')
plt.title('Simulated Field')
plt.axvspan(-10,10,label='Location of sphere',color='orange',alpha=0.2)
plt.legend(bbox_to_anchor=(1.0, 1))
plt.show()
````

The result shows the boundary becoming clearer as the coupled density becomes larger. This is consistent with the expectation.
